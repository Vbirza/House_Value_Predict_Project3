# Population by County

```python
import pandas as pd
```

```python
#Read in the population data 
df = pd.DataFrame(pd.read_csv('../data/Population_Estimate.csv'))

df.head()
```

```python
# Creating a datafram of the population data
df = pd.DataFrame(df)
df.head()
```

```python
# Removing unnecessary columns
pop_clean = df.drop(['Id','April 1, 2010 - Census', 'April 1, 2010 - Estimates Base'], axis=1)
pop_clean.head()
```

```python
# Remaning the columns 
pop_clean = pop_clean.rename(columns={"Population Estimate (as of July 1) - 2010": "2010",
                                      "Population Estimate (as of July 1) - 2011": "2011",
                                      "Population Estimate (as of July 1) - 2012": "2012",
                                      "Population Estimate (as of July 1) - 2013": "2013",
                                      "Population Estimate (as of July 1) - 2014": "2014",
                                      "Population Estimate (as of July 1) - 2015": "2015",
                                      "Population Estimate (as of July 1) - 2016": "2016",
                                      "Population Estimate (as of July 1) - 2017": "2017",
                                      "Population Estimate (as of July 1) - 2018": "2018",
                                     }) 
# Creating new datafram with the cleaned columns    
pop_clean = pop_clean[["Id2", "Geography", "2010", "2011", "2012", "2013","2014", "2015", "2016", "2017", "2018"]]
pop_clean.head()    
```

```python
#Reshape the dataframe so the that "years" columns are transposed into rows grouped by Id2 and Year
pop_stacked = pop_clean.set_index(['Id2', 'Geography']).stack().reset_index().rename({'level_2': 'year', 0: 'pop_chg_yoy'}, axis=1)
pop_stacked.head()
```

```python
#Rearange the datafram so the the ID2 and year are listed first
pop_group = pop_stacked[['Id2', 'year', 'Geography', 'pop_chg_yoy']]
pop_group.head()
```

```python
#Calculated the population percent change for one year 
pc1= pop_group['one_year']=pop_group['pop_chg_yoy'].pct_change(periods=1)

pc1.head()
```

```python
#Calculated the population percent change for two years 
pc2= pop_group['two_years']=pop_group['pop_chg_yoy'].pct_change(periods=2)

pc2.head()
```

```python
#Calculated the population percent change for three years 
pc3= pop_group['three_years']=pop_group['pop_chg_yoy'].pct_change(periods=3)

pc3.head()
```

```python
#Calculated the population percent change for four years
pc4= pop_group['four_years']=pop_group['pop_chg_yoy'].pct_change(periods=4)

pc4.head()
```

```python
#Calculated the population percent change for five years
pc5= pop_group['five_years']=pop_group['pop_chg_yoy'].pct_change(periods=5)

pc5.head()
```

```python
#Final populations change dataframe
pop_group.head()
```

```python
#saving the dataframe to a csv file
pop_group.to_csv('Population_Change.csv')
```
