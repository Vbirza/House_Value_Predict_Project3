# Building Permit Data Pull

```python
import urllib
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser

url = 'https://www2.census.gov/econ/bps/County/'

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(url)
```

```python
#Pulled a listing of all the data text files using beautiful soup HTML parser
```

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
files = []
for item in soup.findAll('td'):
    if len(item.text)==11:
        files.append(item.text)
```

```python
files
```

```python
#for each file in the files array did a URL request and then pulled them into panda dataframe
```

```python
import urllib.request
for file in files:

    urllib.request.urlretrieve(f"https://www2.census.gov/econ/bps/County/{file}", 'permits/{}'.format(file))
```

```python
import pandas as pd

df = pd.DataFrame(pd.read_csv(f'permits/{files[0]}'))
```
