```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv('Population_Estimate.csv'))

df.head()
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
pop_clean = df.drop(['Id','April 1, 2010 - Census', 'April 1, 2010 - Estimates Base'], axis=1)
pop_clean.head()
```

```python
pop_clean = pop_clean.rename(columns={"Population Estimate (as of July 1) - 2010": "2010",
                                      "Population Estimate (as of July 1) - 2011": "2011",
                                      "Population Estimate (as of July 1) - 2012": "2012",
                                      "Population Estimate (as of July 1) - 2013": "2013",
                                      "Population Estimate (as of July 1) - 2014": "2014",
                                      "Population Estimate (as of July 1) - 2015": "2015",
                                      "Population Estimate (as of July 1) - 2016": "2016",
                                      "Population Estimate (as of July 1) - 2017": "2017",
                                      "Population Estimate (as of July 1) - 2018": "2018",
                                     }) 
    
pop_clean = pop_clean[["Id2", "Geography", "2010", "2011", "2012", "2013","2014", "2015", "2016", "2017", "2018"]]
pop_clean.head()    
```

```python
pop_stacked = pop_clean.set_index(['Id2', 'Geography']).stack().reset_index().rename({'level_2': 'year', 0: 'pop_chg_yoy'}, axis=1)
pop_stacked
```

```python
pop_group = pop_stacked[['Id2', 'year', 'Geography', 'pop_chg_yoy']]
pop_group
```
